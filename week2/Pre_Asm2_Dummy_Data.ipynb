```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## [Problem 1] Creating random numbers
Create 500 random numbers with a two-dimensional normal distribution whose mean is (-3, 0) and whose covariance matrix is
$$\begin{bmatrix}
1.0 & 0.8\\
0.8 & 1.0
\end{bmatrix}$$
The shape of the created ndarray will be (500, 2).

```python
np.random.seed(0)

mean = [-3, 0]
cov = [[1.0, 0.8], [0.8, 1.0]]

x1, y1 = np.random.multivariate_normal(mean, cov, 500).T
```

## [Problem 2] Visualization by scatter plot
Visualize the data points created in Problem 1 with a scatter plot.

Scatter plots can be drawn by using matplotlib's `plt.scatter()`.

```python
plt.scatter(x1, y1, marker='o', alpha = 0.5)
plt.title("Scatter")
plt.xlabel("x1")
plt.ylabel("x2")
# plt.axis('equal')
plt.show()
```

## [Problem 3] Visualization by histogram
Visualize the data created in Problem 1 with a histogram.

A histogram can be drawn by using `plt.hist()` 

Create it for each dimension and create two histograms.

The range of the horizontal axis of each histogram should be aligned using `plt.xlim()`.

```python
plt.subplots(figsize=(15, 10))

plt.subplot(221)
plt.hist(x1,50)
plt.xlim(-6,3)

plt.title("Histogram of x1")
plt.xlabel("x1")
plt.ylabel("frequency")

plt.subplot(222)
plt.hist(y1,50)
plt.xlim(-6,3)
plt.title("Histogram of x2")
plt.xlabel("x2")
plt.ylabel("frequency")
```

## [Problem 4] Addition of data
Create 500 new random numbers with a two-dimensional normal distribution whose mean is (0, -3) and whose covariance matrix is represented by the following.
$$\begin{bmatrix}
1.0 & 0.8\\
0.8 & 1.0
\end{bmatrix}$$
Then, visualize the data of problems 1 and 4 as one scatter plot. As a legend, show **0** for Problem 1 and **1** for Problem 4.

```python
np.random.seed(0)

new_mean = [0, -3]
new_cov = [[1.0, 0.8], [0.8, 1.0]]

x2, y2 = np.random.multivariate_normal(new_mean, new_cov, 500).T
```

```python
plt.scatter(x1, y1, marker='o', alpha = 0.5, label = '0')
plt.scatter(x2, y2, marker='o', alpha = 0.5, label = '1')

plt.title("Scatter")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend(loc='best')
plt.show()
```

## [Problem 5] Data combination
It is better to keep the data together so that it is easier to deal with later. Combine the ndarrays created in Problems 1 and 4 into a single ndarray of (1000, 2). 

The concatenation can be done using `np.concatenate()` or `np.vstack()`.

```python
x = np.concatenate((x1,x2))
y = np.concatenate((y1,y2))
xy = np.concatenate((x,y))
combine = np.reshape(xy,(2,1000)).T
print(combine)
print(combine.shape)
```

## [Problem 6] Labeling
For each of the 1000 pieces of data that you have combined together, **label** them to indicate whether they are from question 1 or 4. Question 1 should be labeled **0** and Question 4 should be labeled **1**.


Create a ndarray of (1000, 3) by adding a new column of labels containing 0s and 1s.

Datasets used for machine learning often have this form.

```python
third_col_0 = np.zeros(500)
third_col_1 = np.ones(500)
third_col = np.concatenate((third_col_0,third_col_1))
print(third_col.shape)
```
