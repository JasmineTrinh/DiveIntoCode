# [Problem] The day when chestnut buns cover the solar system

If you sprinkle a single chestnut bun with *Byvine*, how long will it take to cover the solar system with chestnut buns? Create a program to calculate this.

> We assume that the bun is $5$cm in height and $5$cm in radius. So we will use the hemisphere formula to calculate its volume:
$V = \pi h^2 (R - \frac{h}{3}) = \frac{\pi h}{6}(h^2 + 3r^2)\$

> As we know, the Solar system radius is $143.73\times10^9$ km, from the Sun to Sedna, the most distant observable object known in the solar system. Using the sphere formula, we will be able to estimate its volume:
$V = \frac{4}{3}\pi r^3$

```python
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
BUN_VOLUME = math.pi*0.05/6*(0.05**2 + 3*0.05**2)
SOLAR_VOLUME = 4/3*math.pi*143.73*(10**12)
bun_solar_list = []
```

```python
minutes = 0
    
while BUN_VOLUME < SOLAR_VOLUME:
    BUN_VOLUME *= 2
    bun_solar_list.append(BUN_VOLUME)
    minutes += 5
    
print('It will take {:02d} hrs {:02d} mins for the buns to cover the solar system'.format(*divmod(minutes, 60)))
# print(bun_solar_list[-1])
```

```python
plt.title("Coverage of chestnut buns in muinutes to the solar system")
plt.xlabel("minutes")
plt.xlim(0,61)
plt.ylabel("Buns volume (m^3)")
plt.ylim(0,6.5*10**14)
plt.plot(bun_solar_list, 'r--') # Enter the variable name of the list in "List name"
plt.show()
```
